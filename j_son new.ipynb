```python
import requests
import json
import pandas as pd
```

```python
Api_Key='2pDLa5j4iIts8ie0QfcC55oGVCdMcXjV'
response=requests.get('https://api.nytimes.com/svc/mostpopular/v2/viewed/7.json?api-key=2pDLa5j4iIts8ie0QfcC55oGVCdMcXjV').json()
def mdn(response):
    data=[]
    source=[]
    published_date=[]
    adx_keywords=[]
    byline=[]
    title=[]
    abstract=[]
    des_facet=[]
    per_facet=[]
    media=[]


    for i in response ['results']:
        source.append(i['source'])
        published_date.append(i['published_date'])
        adx_keywords.append(i['adx_keywords'])
        byline.append(i['byline'])
        title.append(i['title'])
        abstract.append(i['abstract'])
        des_facet.append(i['des_facet'])
        per_facet.append(i['per_facet'])
        media.append(i['media'])

        x={'source':source,'published_date':published_date,'adx_keywords':adx_keywords,'byline':byline,\
                       'title':title,'abstract':abstract,'des_facet':des_facet, 'per_facet':per_facet,\
                       'media':media}
        data.append(x)    
              

    return data
obj=mdn(response)
```
